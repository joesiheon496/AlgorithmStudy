```python
import pandas as pd
def under_two_reject(df):
    # 1개 이하 제거
    counts = df['hospital_number_id'].value_counts()
    df['annotation_prediction'] = 3
    df = df[df['hospital_number_id'].isin(counts[counts>1].index)]

    # annotation_result 3인것 제거
    df = df[df['annotation_result']!=3]

    # annotation_result nan값 제거
    df.dropna(subset=['annotation_result'],axis=0,inplace=True)

    # 열에 judge,reject list 제거
    df = df.loc[:,~df.columns.str.contains('judge')]
    df['sample_date'] = pd.to_datetime(df['sample_date'])

    reject_list = ['id-2','request_date','id','is_use','annotation-2','annotation_date','annotation','is_imp','age',
                    'gender','hospital_number','id','annotation', 'annotation_date']

    for i in df.columns:
        if i in reject_list:
            df.drop (columns=[f"{i}"],inplace=True)
    return df

# 해당 번호의 환자를 보고 싶다면 해당 코드 사용
def hospital_call_up(df, hospital_number_id):
    # 같은 번호의 환자 아이디 추출
    df = df[df['hospital_number_id']==hospital_number_id].sort_values(by=['sample_date'])
    return df

# A, B, C 리스트 추출
def only_immune_make(df):
    # columns 이름에 judge 제거
    df = df.loc[:,~df.columns.str.contains('judge')]

    reject_list = ['id', 'sample_date''hospital_number_id', 'annotation', 'annotation_date','annotation_result',
               'is_imp','request_date', 'id-2', 'hospital_number', 'age', 'gender', 'is_use', 'annotation-2','sample_date']
    only_immune_make = []

    for i in df.columns:
        if i not in reject_list:
            only_immune_make.append(i)
    return only_immune_make

# 평가
def evalut(df):
    same = 0
    for i,v in zip(df['annotation_result'],df['annotation_prediction']):
        if i == v:
            same +=1
        elif v ==3:
            continue
    count_0 = ((df['annotation_result'] == 0) & (df['annotation_prediction'] == 0)).sum()
    count_1 = ((df['annotation_result'] == 1) & (df['annotation_prediction'] == 1)).sum()
    count_2 = ((df['annotation_result'] == 2) & (df['annotation_prediction'] == 2)).sum()

    print(f"총 개수는 {len(df)}입니다.")
    print(f"0이 {len(df[df['annotation_result'] == 0])}개 중에서 {count_0}개 일치합니다.")
    print(f"1이 {len(df[df['annotation_result'] == 1])}개 중에서 {count_1}개 일치합니다.")
    print(f"2가 {len(df[df['annotation_result'] == 2])}개 중에서 {count_2}개 일치합니다.")

    return same/len(df['annotation_result'])*100
```
